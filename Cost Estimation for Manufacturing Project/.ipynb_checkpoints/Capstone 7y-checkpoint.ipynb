```python
def cleanspaces(text):
    text = re.sub('\s+', '', text)
    return text

def getkw(text):
    patternls = r'(\d{1,5})\s?[Ll][/][Ss]'
    patternkw = r'(\d{1,3})\s?[Kk][Ww]'
    
    flags = re.IGNORECASE
    outputlstext = re.findall(patternls, text, flags = flags)
    outputls = [float(i) for i in outputlstext]
    outputkwtext = re.findall(patternkw, text, flags = flags)
    outputkw = [float(i) for i in outputkwtext]
          
    if len(outputkw) == 0:
        if len(outputls) != 0:
            outputmaxkw = max(outputls)/50
        else:
            outputmaxkw = 0
    else:
        outputmaxkw = max(outputkw)
        
    return outputmaxkw

def getls(text):
    patternls = r'(\d{1,5})\s?[Ll][/][Ss]'
    patternkw = r'(\d{1,3})\s?[Kk][Ww]'
    
    flags = re.IGNORECASE
    outputlstext = re.findall(patternls, text, flags = flags)
    outputls = [float(i) for i in outputlstext]
    outputkwtext = re.findall(patternkw, text, flags = flags)
    outputkw = [float(i) for i in outputkwtext]
    
    if len(outputls) == 0:
        if len(outputkw) != 0:
            outputmaxls = max(outputkw)*50
        else:
            outputmaxls = 0
    else:
        outputmaxls = max(outputls)
        
    return outputmaxls   

def getecon(text):
    patternecon = r'econ'
    
    flags = re.IGNORECASE
    outputecontext = re.findall(patternecon, text, flags = flags)
          
    if len(outputecontext) >= 1:
        outputecon = 1
    else:
        outputecon = 0
       
    return outputecon
```

```python
## Calculate Ridge Regression model

dfall_dummies = pd.get_dummies(data = dfall_clean, columns = ['omlPartGroupID'], 
                                      prefix = ['PG'] )
#Set Target and Features
y = dfall_dummies['cost']
X = dfall_dummies[['kw', 'ls', 'econ','PG_DEHUM', 'PG_ERVCO','PG_PCU' ,'PG_PP', 'PG_RTP' ]]
normalized_X = X.copy()
normalized_X['kw'] = (X['kw']-X['kw'].mean())/X['kw'].std()
normalized_X['ls'] = (X['ls']-X['ls'].mean())/X['ls'].std()

#Split Data Set
X_train, X_test, y_train, y_test = train_test_split(normalized_X, y, test_size=0.2, random_state=42)

# create a model object to hold the modelling parameters
clf = Ridge()

# keep track of the intermediate results for coefficients and errors
coefs = []
errors = []

# create a range of alphas to calculate
ridge_alphas = np.logspace(-6, 6, 200)

# Train the model with different regularisation strengths
for a in ridge_alphas:
    clf.set_params(alpha = a)
    clf.fit(X_train, y_train)
    coefs.append(clf.coef_)
    errors.append(mean_squared_error(clf.coef_, model.coef_))

optimal_ridge = RidgeCV(alphas=ridge_alphas, cv=10)
optimal_ridge.fit(X_train, y_train)
optimal_ridge_alpha = optimal_ridge.alpha_
print('Alpha:', optimal_ridge_alpha)
print('Score:', optimal_ridge.score(X_train, y_train))


y_pred = optimal_ridge.predict(X_test)
RMSE_test =np.sqrt(mean_squared_error(y_test, y_pred))

print(f'RMSE  for Test Data is {RMSE_test}')
```

```python
#define function to take input for year and calculate improved margin

def pred_margin(df, year):
    
    
    startdate = f'{year-1}-6-30'
    enddate = f'{year}-6-30'
  
    mask = (df['completeddate'] > startdate) & (df['completeddate'] <= enddate)
    df = df.loc[mask]
   
    
    df_dummies = pd.get_dummies(data = df, columns = ['omlPartGroupID'], 
                                      prefix = ['PG'] )
    #Set Target and Features
    y = df_dummies['cost']
    X = df_dummies[['kw', 'ls', 'econ','PG_DEHUM', 'PG_ERVCO','PG_PCU' ,'PG_PP', 'PG_RTP' ]]
    normalized_X = X.copy()
    normalized_X['kw'] = (X['kw']-X['kw'].mean())/X['kw'].std()
    normalized_X['ls'] = (X['ls']-X['ls'].mean())/X['ls'].std()

    #apply prediction
    y_pred = model.predict(normalized_X)

    dfall_predict = df.copy()
    dfall_predict['cost_pred'] = y_pred.round(0)
    dfall_predict['GP_pred'] = dfall_predict['sellvalue'] - dfall_predict['cost_pred']
    dfall_predict['GM_pred'] = dfall_predict['GP_pred'] / dfall_predict['sellvalue']
    dfall_predict['GM_pred'] = dfall_predict['GM_pred'].round(2)
    
    h1 = dfall_predict
    h2 = dfall_predict
    
    dfall_predicthm = dfall_predict[dfall_predict['GM_pred'] > 0.25]

    h1hm = dfall_predicthm
    h2hm = dfall_predicthm
    
    print(f'Actual Results for FY: {year}')
    print(f'{startdate} to {enddate}')
    print(f'______________________________________')
    print(f'Actual Mean GM: {h1["GM"].mean()}\nPredicted Mean GM: {h2["GM_pred"].mean()}')
    print(f'Actual Median GM: {h1["GM"].median()}\nPredicted Median GM: {h2["GM_pred"].median()}')
    print(f'Actual Total GP: {h1["GP"].sum()}\nPredicted Total GP: {h2["GP_pred"].sum()}')
    print(f'Actual Total Cost: {h1["cost"].sum()}\nPredicted Total Cost: {h2["cost_pred"].sum()}')
    print(f'Actual Total Revenue: {h1["sellvalue"].sum()}')
    print(f'Actual Total Labour Hours: {h1["labourhours"].sum()}')
    print(f'Actual GM: {h1["GP"].sum()/h1["sellvalue"].sum()}')
    print(f'______________________________________')
    print(f'\nNew results for FY: {year} with GM > 25%')
    print(f'______________________________________')
    print(f'New Mean: {h1hm["GM"].mean()}\nPredicted Mean: {h2hm["GM_pred"].mean()}')
    print(f'New Median: {h1hm["GM"].median()}\nPredicted Median: {h2hm["GM_pred"].median()}')
    print(f'New Total GP: {h1hm["GP"].sum()}\nPredicted Total GP: {h2hm["GP_pred"].sum()}')
    print(f'New Total Cost: {h1hm["cost"].sum()}\nPredicted Total Cost: {h2hm["cost_pred"].sum()}')
    print(f'New Total Revenue: {h1hm["sellvalue"].sum()}')
    print(f'New Total Labour Hours: {h1hm["labourhours"].sum()}')
    print(f'New GM: {h1hm["GP"].sum()/h1hm["sellvalue"].sum()}')
    print(f'______________________________________')
    print(f'\nDeltas')
    print(f'______________________________________')
    print(f'Delta GP: {h1hm["GP"].sum()-h1["GP"].sum()}')
    print(f'Delta Cost: {h1hm["cost"].sum()-h1["cost"].sum()}')
    print(f'Delta Revenue: {h1hm["sellvalue"].sum()-h1["sellvalue"].sum()}')
    print(f'Delta Labour Hours: {h1hm["labourhours"].sum()-h1["labourhours"].sum()}')
    print(f'Delta GM: {(h1hm["GP"].sum()/h1hm["sellvalue"].sum())-(h1["GP"].sum()/h1["sellvalue"].sum())}')
    print(f'______________________________________')
    
    #yearlist is [year, total labour hours,  saving in labour hours, actual gross profit, new gross profit, new gross margin ]
    yearlist = [year, h1["labourhours"].sum() ,h1hm["labourhours"].sum()-h1["labourhours"].sum(), h1["GP"].sum(), h1hm["GP"].sum(), h1hm["GP"].sum()/h1hm["sellvalue"].sum() ]
    return yearlist

```
