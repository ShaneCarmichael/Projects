## Define functions

```python
def printhist(dfcol,xlabel = 'x',title = 'Title',color = 'blue'):#print histogram
    median = (f'Median = {round(dfcol.median(),2)}')
    plt.figure()
    plt.rcParams["axes.labelsize"] = 15
    plt.rcParams["axes.titlesize"] = 20
    ax = dfcol.hist(bins = 40, figsize = (10,8), color = color, alpha = 0.8)
    plt.xlabel(xlabel)
    plt.ylabel('Frequency')
    plt.title(title)
    at = AnchoredText(median,frameon=True,loc='upper right',)
    at.patch.set_boxstyle("round,pad=0.,rounding_size=0.2")
    ax.add_artist(at)
    plt.show()

#print two histograms 
def printhist2(dfcol1, dfcol2,dfcol1_label = 'df1',dfcol2_label = 'df2',xlabel = 'x',title = 'Title',color1 = 'orange'
              , color2 = 'green'):
    plt.figure()
    plt.rcParams["axes.labelsize"] = 15
    plt.rcParams["axes.titlesize"] = 20

    dfcol1.hist(bins = 80, label = dfcol1_label, density = True, figsize = (10,8), color = color1, alpha = 0.8)
    dfcol2.hist(bins = 80,label = dfcol2_label,  density = True, figsize = (10,8), color = color2, alpha = 0.8)
    plt.xlabel(xlabel)
    plt.ylabel('Frequency')
    plt.title(title)
    plt.legend()
    plt.show()
    
#print line plot
def printlineplot(x,y,xlabel = 'x',ylabel = 'y',title = 'Title',color = 'blue'):
    plt.figure()
    plt.rcParams["axes.labelsize"] = 15
    plt.rcParams["axes.titlesize"] = 20
    ax = plt.plot(x,y,color = color, alpha = 0.8)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.show()
    
def mean_absolute_percentage_error(y_true, y_pred): 
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

#normalise data and get dummies
def norm(df):#normalise data
    df_dummies = pd.get_dummies(data = df, columns = ['omlPartGroupID'],prefix = ['PG'] )
    #Set Target and Features
    y = df_dummies['cost']
    X = df_dummies[['kw', 'ls', 'econ' ,'PG_DEHUM', 'PG_ERVCO','PG_PCU' ,'PG_PP', 'PG_RTP', 'age']]
    #normalise x
    normalized_X = X.copy()
    #normalized_X = (X-X.mean())/X.std()
    normalized_X['kw'] = (X['kw']-X['kw'].mean())/X['kw'].std()
    normalized_X['ls'] = (X['ls']-X['ls'].mean())/X['ls'].std()
    normalized_X['age'] = (X['age']-X['age'].mean())/X['age'].std()
    return normalized_X, y

def pca(X, n_components):#perform PCA and plt explained variance
    pca = PCA()
    pca.fit(X)
    expv = pca.explained_variance_
    xplot = []  
    pca_new = PCA(n_components=n_components)
    transformedx = pca_new.fit_transform(X)
    transformedx_dataframe = pd.DataFrame(transformedx) 
    for i in range(len(expv)):
        xplot.append(i+1)
    plt.plot(xplot, expv)
    plt.title('PCA Features',size = 15)
    return transformedx_dataframe

def linregmodel(df, ispca=False, n_components = 1):#pick dataframe and pca if applicable
    normalized_X, y = norm(df)
    #if PCA is chosen
    if ispca == True:
        X_actual = pca(normalized_X, n_components)
    else:
        X_actual = normalized_X
        n_components = 'NA'
    model = LinearRegression()
    X_train, X_test, y_train, y_test = train_test_split(X_actual, y, test_size=0.2, random_state=42)
    model.fit(X_train, y_train)
    R2_train = model.score(X_train, y_train).round(2)
    R2_test = model.score(X_test, y_test).round(2)
    y_pred = model.predict(X_test).round(2)
    RMSE_test =np.sqrt(mean_squared_error(y_test, y_pred)).round(2)
    MAPE_test = mean_absolute_percentage_error(y_pred, y_test).round(2)
    #create dataframe of results
    df_results = pd.DataFrame([{'Model Type':model,'R2 Train Score':R2_train, 'R2 Test Score':R2_train, 
                               'RMSE Test':RMSE_test, 'MAPE Test %':MAPE_test, 'PCA':ispca, 'n':n_components}])
    return df_results
    plt.show()

def anymodel(model, df, ispca=False, n_components = 1):#pick model, dataframe and pca if applicable
    normalized_X, y = norm(df)

    #if PCA is chosen
    if ispca == True:
        X_actual = pca(normalized_X, n_components)
    else:
        X_actual = normalized_X
        n_components = 'NA'
    model = model
    X_train, X_test, y_train, y_test = train_test_split(X_actual, y, test_size=0.2, random_state=42)
    model.fit(X_train, y_train)
    y_pred_train = model.predict(X_train)
    y_pred_test = model.predict(X_test)
    
    R2_Train_Score = r2_score(y_train,y_pred_train)
    R2_Test_Score = r2_score(y_test,y_pred_test)
    
    
    RMSE_test =np.sqrt(mean_squared_error(y_test, y_pred_test)).round(2)
    MAPE_test = mean_absolute_percentage_error(y_pred_test,y_test).round(2)

    #create dataframe of results
    df_results = pd.DataFrame([{'Model Type':model,'R2 Train Score':R2_Train_Score, 'R2 Test Score':R2_Test_Score, 
                               'RMSE Test':RMSE_test, 'MAPE Test %':MAPE_test, 'PCA':ispca, 'n':n_components}])
    return df_results
    plt.show()

def predmargin(model, df, startdate, enddate, ispca=False, n_components = 1,minimumgm = 0.25):#predict and get margins
   
    normalized_X, y = norm(df)
    
    #if PCA is chosen
    if ispca == True:
        X_actual = pca(normalized_X, n_components)
    else:
        X_actual = normalized_X
        n_components = 'NA'
    model = model
    X_train, X_test, y_train, y_test = train_test_split(X_actual, y, test_size=0.2, random_state=42)
    model.fit(X_train, y_train)
    
    startdate = startdate
    enddate = enddate
    mask = (df['completeddate'] > startdate) & (df['completeddate'] <= enddate)
    df = df.loc[mask]
    X_actual_filter = X_actual.loc[mask]
    y_pred = model.predict(X_actual_filter)
    
    df_predict = df.copy()
    df_predict['cost_pred'] = y_pred.round(0)
    df_predict['GP_pred'] = df_predict['sellvalue'] - df_predict['cost_pred']
    df_predict['GM_pred'] = df_predict['GP_pred'] / df_predict['sellvalue']
    df_predict['GM_pred'] = df_predict['GM_pred'].round(2)
    
    df_predicthm = df_predict[df_predict['GM_pred'] > minimumgm]
    
    labour_actual = df_predict["labourhours"].sum()
    revenue_actual = df_predict["sellvalue"].sum()
    cost_actual = df_predict["cost"].sum()
    GP_actual = revenue_actual - cost_actual
    GM_actual = (GP_actual / revenue_actual)
    
    labour_hm = df_predicthm["labourhours"].sum()
    revenue_hm = df_predicthm["sellvalue"].sum()
    cost_hm = df_predicthm["cost"].sum()
    GP_hm = revenue_hm - cost_hm
    GM_hm = GP_hm / revenue_hm
    
    Revenuedrop = revenue_actual - revenue_hm
    costsaving = cost_actual - cost_hm
    laboursaving = labour_actual - labour_hm
    increasedcapacity = (laboursaving / labour_hm)*100
    increasedgp = (increasedcapacity)*GP_hm/100
    
    df_results_actual = pd.DataFrame([{'Results':'Actual','Labour Hours':labour_actual,'Revenue':revenue_actual, 'Cost':cost_actual, 
                               'Gross Profit':GP_actual, 'Gross Margin':GM_actual}])
    df_results_hm = pd.DataFrame([{'Results':'Filtered by Margin','Labour Hours':labour_hm,'Revenue':revenue_hm, 'Cost':cost_hm, 
                               'Gross Profit':GP_hm, 'Gross Margin':GM_hm}])
    df_results1 = (pd.concat([df_results_actual,df_results_hm])).round(2)
    
    df_results2 = pd.DataFrame([{'Start Date':startdate,'End Date':enddate,'GM Threshold':minimumgm,
                                 'Revenue Drop':Revenuedrop,'Cost Saving':costsaving, 'Labour Saving': laboursaving,
                                 'Additional Prod Cap %':increasedcapacity,'Additional GP':increasedgp}]).round(2)
    
    return df_results1, df_results2, df_predict, df_predicthm


def optimalgm(model, df, startdate, enddate, ispca=False, n_components = 1):#print optimal margin chart
    df_combined = pd.DataFrame()
    for i in range(30):
        i = i+1
        j = i/100
        temp1, temp2, temp3, temp4 = predmargin(model, df, startdate, enddate, ispca=ispca, n_components = n_components,minimumgm = j)
        df_combined = pd.concat([df_combined,temp2])

    # create figure and axis objects with subplots()
    fig,ax = plt.subplots(figsize=(10,8))
    # make a plot
    ax.plot(df_combined['GM Threshold'], df_combined['Additional GP'], color="green")
    # set x-axis label
    ax.set_xlabel("Minimum Gross Margin Threshold",fontsize=14)
    # set y-axis label
    ax.set_ylabel("Increase in Gross Profit $",color="green",fontsize=14)
    ax.set_axisbelow(True)
    ax.yaxis.grid(color='gray', linestyle='dashed')

    # twin object for two different y-axis on the sample plot
    ax2=ax.twinx()
    # make a plot with different y-axis using second axis object
    ax2.plot(df_combined['GM Threshold'], df_combined['Additional Prod Cap %'],color="black")
    ax2.set_ylabel("Additional Sales % \n Needed to Fill Production", color="black",fontsize=14)
    ax2.set_axisbelow(True)
    ax2.yaxis.grid(color='gray', linestyle='dashed')
    plt.show()
    
    
def calculate_age(datecompleted):#get age
    #datecompleted = datetime.strptime(datecompleted, "%d.%m.%Y").date()
    today = date.today()
    #return today.year - datecompleted.year - ((today.month, today.day) < (datecompleted.month, datecompleted.day))
    return today - datecompleted

def ann(df, epochs = 5):# Neural Networks model
    normalized_X, y = norm(df)
    #split data
    X_train, X_test, y_train, y_test = train_test_split(normalized_X, y, test_size=0.2, random_state=42)

    #input shape
    n_cols = X_train.shape[1]
    #create model
    model = Sequential()
    model.add(Dense(100, activation = 'relu', input_shape = (n_cols, )))
    model.add(Dense(100, activation = 'relu'))
    model.add(Dense(50, activation = 'relu'))
    model.add(Dense(1, activation='linear'))
    
    #compile model
 
    model.compile(optimizer = 'adam', loss = 'mean_squared_error',metrics = ['mean_squared_error'])
    print(model.summary())
    
    model.fit(X_train, y_train, epochs=epochs, validation_split=0.2, verbose = False)
    
    y_pred_train = model.predict(X_train)
    y_pred_test = model.predict(X_test)
    
    R2_Train_Score = r2_score(y_train,y_pred_train)
    R2_Test_Score = r2_score(y_test,y_pred_test)
    
    
    RMSE_test =np.sqrt(mean_squared_error(y_test, y_pred_test)).round(2)
    MAPE_test = mean_absolute_percentage_error(y_pred_test,y_test).round(2)

    #create dataframe of results
    df_results = pd.DataFrame([{'Model Type':model,'R2 Train Score':R2_Train_Score, 'R2 Test Score':R2_Test_Score, 
                               'RMSE Test':RMSE_test, 'MAPE Test %':MAPE_test, 'PCA':False, 'n':'NA'}])
    return df_results
```

```python
dflinreg = anymodel(LinearRegression(),df_clean, False, 5)
dfridge = anymodel(Ridge(alpha = 0.9),df_clean, False, 3)
dflasso = anymodel(Lasso(alpha = 0.45),df_clean, False, 3)
dfagginglinreg = anymodel(BaggingRegressor(base_estimator = LinearRegression(),n_estimators = 100,max_samples = 0.8,max_features = 0.8),df_clean, False, 3)
dfadaboost = anymodel(AdaBoostRegressor(random_state=0, n_estimators=100),df_clean, False, 3)
dfxgboost = anymodel(XGBRegressor(n_estimators=1000, max_depth=7, eta=0.1, subsample=0.7, colsample_bytree=0.8),df_clean, False, 3)
dfrandomforest = anymodel(RandomForestRegressor(n_estimators = 1000, random_state = 42),df_clean, False, 4)
dfann = ann(df_clean, 200)
eval_df = pd.concat([dflinreg,dfridge, dflasso, dfagginglinreg, dfadaboost, dfxgboost, dfrandomforest, dfann])
model_list = ['Linear Regression', 'Ridge', 'Lasso', 'Bagging with LR', 'AdaBoost', ' XGBoost', 'Random Forest', 'Deep Learning']
eval_df['Model Name'] = model_list 
col_list = [7,1,2,3,4]
eval_df_present = eval_df.iloc[:,col_list]
eval_df_present
```
